```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import pandas as pd
from scipy.optimize import curve_fit

plt.style.use('bmh')
```

```python
def f(x, a, b):
    return a*x + b

def plot_and_extrapolate(x,y):
    plt.plot(x, y, label = 'data')

    (a, b), cov = curve_fit(f, x, y)

    plt.plot(x, [f(i, a, b) for i in x], '--', label='fit')
    plt.legend()

    print(f'The extrapolated intercept is at the value {-b/a}')
    return -b/a
```

```python
# first looking at above data

df = pd.read_csv('pc_sc_test_from_above')
df.columns = ['x', 'y']
x = np.array(df['x'])[10:]
y = np.array(df['y'])[10:]

plot_and_extrapolate(x,y)
```

```python
# Now at data below

df = pd.read_csv('pc_sc_test_from_below')
df.columns = ['x', 'y']
x = np.array(df['x'])[10:]
y = np.array(df['y'])[10:]

expected = plot_and_extrapolate(x,y)
```

## Values of the intercept agree to within 5 decimal places

#### after running at the expected value, the error is difference per 2000 runs is 4 orders of magnitude less
